```python
# Descriptive Analysis function
def Descriptive_analysis (data):
    stats_table = data.describe()

    stats_table['variance'] = data.var()
    stats_table['std_dev'] = data.std()
    stats_table['range'] = data.max() - data.min()
    stats_table['IQR'] = stats_table['75%'] - stats_table['25%']
    stats_table['skewness'] = data.skew()
    stats_table['kurtosis'] = data.kurtosis()
    pd.options.display.float_format = '{:.2f}'.format
    display(pd.DataFrame(stats_table))
    
    plt.figure(figsize=(15, 6))
    
    # Boxplot
    plt.subplot(1, 2, 1)
    sns.boxplot(x=data, color='lightblue', fliersize=5)
    plt.title('Boxplot of GDP')
    
    # Histogram
    plt.subplot(1, 2, 2)
    sns.histplot(data, bins=20, kde=True, color='lightblue', edgecolor='black')
    plt.title('Histogram of Closing price')
    plt.xlabel('Price')
    plt.ylabel('Frequency')

    # Hiển thị biểu đồ
    plt.tight_layout()
    plt.show()
```

```python
# SSI stock
Descriptive_analysis(SSI['Close'])
```

```python
# VND stock
Descriptive_analysis(VND['Close'])
```

```python
# VIB stock
Descriptive_analysis(VIB["Close"])
```
