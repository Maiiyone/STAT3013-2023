```python
# 1. Các thư viện cần thiết
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, OneHotEncoder,StandardScaler,PowerTransformer, MinMaxScaler, RobustScaler
from sklearn.model_selection import KFold, cross_val_predict, train_test_split,GridSearchCV,cross_val_score

from sklearn.linear_model import LinearRegression, Lasso, Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# 2. Đọc file dữ liệu
df = pd.read_csv('STB.csv')
df['Date'] = pd.to_datetime(df['Date'])
train_dates= pd.to_datetime(df['Date'])
df.head()
len(df)
```

```python
# Xem data 
df# Split data into training (70%) and test (30%)
train_data, test_data = train_test_split(df, test_size=0.3, shuffle=False)
```

```python
# Drop the 'Date' column from both X_train and X_test
X_train = train_data.drop(['Price', 'Date'], axis=1)
y_train = train_data['Price']

X_test = test_data.drop(['Price', 'Date'], axis=1)
y_test = test_data['Price']

model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
n_days_for_prediction= 30
cols = list(df)[1:4]
predict_period_dates = pd.date_range(list(train_dates)[-1], periods=n_days_for_prediction).tolist()
print(predict_period_dates)

predict_period_dates_index = pd.RangeIndex(start=len(df),stop=len(df)+30,step=1)

#Make prediction
prediction1 = model.predict(X_test[-n_days_for_prediction:])
df_for_training = df[cols].astype(float)
print(prediction1)
```

```python
# 6. Vẽ hình
plt.figure(figsize=(10, 6))
plt.plot(test_data.index, test_data['Price'], linestyle='-', color='r', label='Test Data (Actual)')
plt.plot(train_data.index, train_data['Price'], linestyle='-', color='b', label='Training Data (Actual)')
plt.plot(test_data.index, y_pred, linestyle='-', color='orange', label='Test Data (Predicted)')
plt.plot(predict_period_dates_index, prediction1, linestyle='-', color='g', label='Predict for 60 next days')
plt.title('Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
# 7. Đánh giá độ chính xác test
test_rmse = np.sqrt(np.mean((y_pred - y_test)**2))
test_mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
test_mae = mean_absolute_error(y_test, y_pred)

print('Testing RMSE:', test_rmse)
print('Testing MAPE:', test_mape)
print('Testing MAE:', test_mae)
```
